### Data Cleaning and manipulation

```python
# Let's make the dataframe columns titles suitable for python. 
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
# Splitting the inspection_type column into separate columns as there are 2 different values in it.
df['inspection_type_1'] = df['inspection_type'].str.split(' /', expand = True)[0]   
df['inspection_type_2'] = df['inspection_type'].str.split(' /', expand = True)[1]
```

```python
df.drop(['inspection_type'] , axis = 1 , inplace = True )   # Dropping the column inspection_type 
```

```python
df['inspection_date'] = pd.to_datetime(df['inspection_date']) # Changing Inspection date column to datetime format 

# Seperating the Year , Month and the day of the week out from the inspection_date column.
df['inspection_year'] = df['inspection_date'].dt.year
df['inspection_month'] = df['inspection_date'].dt.month
df['day_of_the_week'] = df['inspection_date'].dt.weekday_name
```

```python
df.drop(['inspection_date'] , axis = 1 , inplace = True )   # Dropping the inspection_date column
```

# Imputing missing values in our dataframe

```python
# First, Let's create a new dataframe with the selected features and our target feature from the dataset.

new_df = df[['boro' , 'inspection_year' , 'inspection_month' ,  
             'day_of_the_week' , 'cuisine_description' , 
             'score' , 'inspection_type_1' , 'inspection_type_2','longitude', 'latitude',  'critical_flag']].copy()
```

```python
from scipy.stats import iqr
iqr=iqr(new_df['score'])
iqr
```

```python
# Finding the upper and lower bound of the score variable

price_stats=new_df['score'].describe()
upper_bound=price_stats['75%']+(1.5*iqr)
lower_bound=price_stats['25%']-(1.5*iqr)
outliers={'above_upper':0,'below_lower':0}
indexes=[]
for i,j in enumerate(new_df['score'].values):
    if j>upper_bound:
        outliers['above_upper']+=1
        indexes.append(i)
    elif j<lower_bound:
        outliers['below_lower']+=1
        indexes.append(i)
print (f" Upper Bound: {upper_bound}  \n Lower bound : {lower_bound} ")
#print(f"{val}for{val} is a portal for {val}.")
```

## Converting categorical features to numerical: 

```python
# encoding yes as 1 and no as 0 for critical flag variable
new_df['critical_flag'] = new_df['critical_flag'].replace(to_replace = 'N', value=0)
new_df['critical_flag'] = new_df['critical_flag'].replace(to_replace = 'Y', value=1)
```

```python
#Encoding Days as numbers with Monday as 0,....,Sunday as 6

from datetime import datetime
import time
new_df['day_of_the_week']=new_df['day_of_the_week'].apply(lambda x: time.strptime(x, '%A').tm_wday)
```

```python
# Converting categorical variables into dummy variables.

#pd.get_dummies(df,['purpose'],drop_first=True)

BORO_dummy_df=pd.get_dummies(new_df['boro'],prefix='boro' , drop_first=True )
INSPECTION_TYPE1_dummy_df=pd.get_dummies(new_df['inspection_type_1'],prefix='it1' , drop_first=True)
INSPECTION_TYPE2_dummy_df=pd.get_dummies(new_df['inspection_type_2'],prefix='it2' , drop_first=True)
#ACTION_dummy_df=pd.get_dummies(new_df['action'],prefix='action' , drop_first=True)
```

```python
# Encoding the cuisine_description column using One hot encoding.
new_df['cuisine_description'] = new_df['cuisine_description'].astype('category').cat.codes 
```

```python
# Combining the dataframe and the dummy variables in a new data frame.
new_df_1 = pd.concat([new_df,BORO_dummy_df, INSPECTION_TYPE1_dummy_df, 
                      INSPECTION_TYPE2_dummy_df] , axis = 1)
```

```python
new_df_1.drop (['boro' , 'cuisine_description' , 'inspection_type_1' ,
                   'inspection_type_2' ] , axis =1 , inplace= True)
```
